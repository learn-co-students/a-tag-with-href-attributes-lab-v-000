# **☁️Word Cloud☁️**

```python
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords

plt.figure(figsize = (10,10))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(
                      background_color = 'black',
                      stopwords = stopwords,
                      max_words = 1000,
                      max_font_size = 120,
                      random_state = 42
                    ).generate(str(data['Review Text']))

#Plotting the word cloud
plt.imshow(wordcloud)
plt.title("WORD CLOUD for Review Text", fontsize = 15)
plt.axis('off')
plt.show()
```

"Really" and "Fit" were the most used words across all of our reviews. Excluding "Really", this shows that our customers are most concerned with how things fit.
